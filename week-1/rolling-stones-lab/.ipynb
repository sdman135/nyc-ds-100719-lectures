
### Functions to build-out:

Each of the following functions can be defined in the `functions.py` file. 

* **Searching functions**
  * Find by name - Takes in a string that represents the name of an album. Should return a dictionary with the correct album, or return `None`.
  * Find by rank - Takes in a number that represents the rank in the list of top albums and returns the album with that rank. If there is no album with that rank, it returns `None`.
  * Find by year - Takes in a number for the year in which an album was released and returns a list of albums that were released in that year. If there are no albums released in the given year, it returns an empty list.
  * Find by years - Takes in a start year and end year. Returns a list of all albums that were released on or between the start and end years. If no albums are found for those years, then an empty list is returned. 
  * Find by ranks - Takes in a start rank and end rank. Returns a list of albums that are ranked between the start and end ranks. If no albums are found for those ranks, then an empty list is returned.
* **All functions**
  * All titles - Returns a list of titles for each album.
  * All artists - Returns a list of artist names for each album.
* **Questions to answer / functions**
  * Artists with the most albums - Returns the artist with the highest amount of albums on the list of top albums 
  * Most popular word - Returns the word used most in amongst all album titles
  * Histogram of albums by decade - Returns a histogram with each decade pointing to the number of albums released during that decade.
  * Histogram by genre - Returns a histogram with each genre pointing to the number of albums that are categorized as being in that genre.

````python
""" Find by name - Takes in a string that represents the name of an album. Should return a dictionary with the correct album, or return None. """
def find_by_name(name):
    
    #looks for album name and displays it's Dictionary
    
    for album in info:                    # running through our info list and labeling each element in it album
        if album['album'] == name:        # only if ['album'] key matches the name given in function (find_by_name(?))
            return album                  # returns the values we stored as albums
        
 
````

````python
""" Find by rank - Takes in a number that represents the rank in the list of top albums and returns the album with that rank. If there is no album with that rank, it returns None. """
def find_by_rank(rank):
    
    #looks for the rank and displays it's album name 
    
    for number in info:                 # iterating through our info and labeling each element number
        if number['number'] == rank:    # only if  the element's ['number'] key matches the parameter called in function(in find_by_name(?))
            return number['album']      # then we will returns values of the 'rank' given(ie ['number'])
        else:
            return None
    
       
````

````python
# Find by year - Takes a number for the year in which an album was released and returns a list of albums that were released in that year. #

def find_by_year(year):
    released_same_year = []                               # Made an empty list for albums released in the same year
    for number in info:
        if number['year'] == year:
            released_same_year.append(number['album'])    # Appending to the empty list the results to "released_same_year = []"
    return released_same_year
    
find_by_year('1965')
````

````python
# Find by ranks - Takes in a start rank and end rank. Returns a list of albums that are ranked between the start and end ranks. If no albums are found for those ranks, then an empty list is returned.# 

def find_by_ranks(start_rank, end_rank): ## Very similar to find_by_years (replaced [years] with [ranks])
    
    group_range_of_ranks =[]                                    ### empty list to the collection of albums at a range of ranges
    for number in range(int(start_rank),int(end_rank)+1):       ### the +1  is needed b/c if not the range will not display the correct ending range
        group_range_of_ranks.append(find_by_rank(str(number)))  ### if you use append it will make a list of each first character
    return group_range_of_ranks
        
````

````python
# Artists with the most albums - Returns the artist with the highest amount of albums on the list of top albums #

def artist_with_most_albums(): 

    # Create and populate frequency distribution
   
    frequency_dict = {}                       ### Empty frequency count list
    #albums = all_artist()                    ### realized i dont need this can iterate directly in next line of code
    
    # For all elements in the list:
    for item in all_artist():                 ### replaced "albums" with "all_artist()" eliminated the whole line of code before

    
        if item not in frequency_dict:        # If an element is not in the dictionary, add it with value 1
            frequency_dict[item] = 1

    
        else:
                frequency_dict[item] += 1     # If an element is already in the dictionary, +1 the value
   
    # Create a list for mode values
    mode = []
    max_value = max(frequency_dict.values())  # making the highest value(ie count) the largest(ie max) frequency
    for key, value in frequency_dict.items(): 
        if value == max_value:
            mode.append({key:value})          #adds the artist{key} with the largest frequency and the total number of albums count{value}
    return mode
````
